#Preprocessing

```python
# Separate numerical and categorical columns
numerical_columns = data.select_dtypes(include=[np.number]).columns.tolist()
categorical_columns = data.select_dtypes(include=['object']).columns.tolist()
```

```python
# Handle missing values:
# - For numerical columns, use median imputation.
# - For categorical columns, use mode imputation.
data[numerical_columns] = data[numerical_columns].fillna(data[numerical_columns].median())
for col in categorical_columns:
    data[col] = data[col].fillna(data[col].mode()[0])
```

```python
# Remove duplicates or redundant rows
data.drop_duplicates(inplace=True)
```

```python
# Standardize and unify column names for consistency
data.columns = data.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# Encoding Categorical Variables using Label Encoding
label_columns = ['gender', 'education', 'country', 'ethnicity']
label_encoder = LabelEncoder()
for col in label_columns:
    data[col] = label_encoder.fit_transform(data[col])
```

```python
# List of drug columns
drug_columns = ['alcohol',
       'amphet', 'amyl', 'benzos', 'caff', 'cannabis', 'choc', 'coke', 'crack',
       'ecstasy', 'heroin', 'ketamine', 'legalh', 'lsd', 'meth', 'mushrooms',
       'nicotine', 'semer', 'vsa']

# Count occurrences of each class (0s and 1s) for each drug
for drug in drug_columns:
    print(f"\n{drug}:\n", data[drug].value_counts())
```

```python
# Encode drug usage (Binary Classification)
drug_columns = [
      'amphet', 'amyl', 'benzos', 'cannabis','coke',
       'ecstasy', 'ketamine', 'legalh', 'lsd', 'meth', 'mushrooms',
       'nicotine','vsa']


for drug in drug_columns:
    data[drug] = data[drug].replace({'CL0': 0, 'CL1': 0, 'CL2': 1, 'CL3': 1, 'CL4': 1, 'CL5': 1, 'CL6': 1})

# Create a new binary target variable
data['drug_use'] = data[drug_columns].max(axis=1)

# Check class distribution
print(data['drug_use'].value_counts())
```

##Dimensionality Reduction using PCA

Principal Component Analysis (PCA) reduces the dimensionality of your data while retaining the maximum variance.

Steps:

Standardize the data before applying PCA.
Determine the number of components that explain at least 95% of the variance.

```python
# Apply PCA with 2 components (for visualization)
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_scaled)
```

```python
# Scatter plot of PCA results
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=y, cmap='crest', edgecolor='k', s=60, alpha=0.7)
plt.colorbar(scatter, label='Drug Consumption (0=No, 1=Yes)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Personality Traits (First 2 Components)')
plt.show()
```

```python
# Variance explained bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=['PC1', 'PC2'], y=explained_variance_2d, palette='mako')
plt.title('Variance Explained by PCA Components (2D)')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

##Oversampling with SMOTE

SMOTE (Synthetic Minority Oversampling Technique) generates synthetic samples for the minority class to balance the dataset.

```python
# Split the data first
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Step 3: Apply Balancing Techniques
# Set only ONE to True
apply_smote = True
apply_random_oversampler = False
apply_adasyn = False
apply_nearmiss = False
apply_random_undersampler = False
apply_tomeklinks = False
```

```python
# Balancing function
def balance_data(X, y, method=None):
    if method == 'smote':
        sampler = SMOTE(random_state=42, k_neighbors=1)
    elif method == 'adasyn':
        sampler = ADASYN(random_state=42, n_neighbors=1)
    elif method == 'random_over':
        sampler = RandomOverSampler(random_state=42)
    elif method == 'random_under':
        sampler = RandomUnderSampler(random_state=42)
    elif method == 'nearmiss':
        sampler = NearMiss()
    elif method == 'tomeklinks':
        sampler = TomekLinks()
    else:
        return X, y  # No sampling
    return sampler.fit_resample(X, y)
```

```python
# Determine which method to apply
resample_method = None
if apply_smote:
    resample_method = 'smote'
elif apply_adasyn:
    resample_method = 'adasyn'
elif apply_random_oversampler:
    resample_method = 'random_over'
elif apply_random_undersampler:
    resample_method = 'random_under'
elif apply_nearmiss:
    resample_method = 'nearmiss'
elif apply_tomeklinks:
    resample_method = 'tomeklinks'
```

```python
# Apply resampling
X_train_res, y_train_res = balance_data(X_train, y_train, resample_method)
```

```python
# Step 4: Show Class Distribution After Balancing
print(f"\n Class Distribution After {resample_method.upper()}:\n")
print(y_train_res.value_counts())
print("\nClass Distribution (%):\n", y_train_res.value_counts(normalize=True) * 100)
plot_class_distribution(y_train_res, f"Class Distribution After {resample_method.upper()}")
```
